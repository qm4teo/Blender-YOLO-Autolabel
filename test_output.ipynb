## Imports

```python
import cv2
import matplotlib.pyplot as plt
from test_utils import draw_bounding_boxes, parse_file

%matplotlib inline
```

## Single image

```python
bboxes = parse_file('labels/0001.txt')
img = cv2.imread('images/0001.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

draw_bounding_boxes(img, bboxes, color=(255, 0, 0))
```

## Image set

```python
image_set = "A"

for i in range(1,6):
    bboxes = parse_file(f'labels/{image_set}_{i:04d}.txt')
    img = cv2.imread(f'images/{image_set}_{i:04d}.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    print(f'{image_set}_{i:04d}.jpg')
    draw_bounding_boxes(img, bboxes)
```
